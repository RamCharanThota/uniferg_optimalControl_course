```python
import autograd.numpy as np
from autograd import grad , jacobian
import matplotlib.pyplot as plt
import control
```

```python
# dynamics of the system
def dynamics(x,u):
   dx=((1+x)*x)+2*u
   return dx

# Define the Runge-Kutta 4th order integration function to get the next step
def rk4_step( x, u, dt):
    k1 = dt*dynamics(x,u)
    k2 = dt*dynamics(x + k1/2, u)
    k3 = dt*dynamics(x + k2/2, u)
    k4 = dt*dynamics(x + k3, u)
    return x + (1/6.0) * (k1 + 2*k2 + 2*k3 + k4)

dt=0.1

x_equ=0.0
u_equ=0.0

Ad=jacobian(rk4_step,argnum=0)(x_equ,u_equ,dt)
Bd=jacobian(rk4_step,argnum=1)(x_equ,u_equ,dt)

#Discretesized non linear system f(x) using RK45
print("the state matrix:",Ad,"the control matrix",Bd)


```

```python
#cost function
Q=2
R=1
QN=2

# solution to the 2.1 
S,eig,K=control.dare(Ad,Bd,Q,R)
print(" the cost to go P value is:",S ," the gain K value is :",K)

A=Ad
B=Bd
```

```python
# backward loop to calculate cost to go and control gain using dynamic programming
dt=0.1
nSteps=20
t_span=np.linspace(0,dt*nSteps,nSteps+1)

# Xn would be our goal
xgoal=0.2;

Sk=np.zeros(nSteps+1)
Uk=np.zeros(nSteps+1)
Kk=np.zeros(nSteps+1)
Vk=np.zeros(nSteps+1)

Xk=np.zeros(nSteps+1)


Sk[20]=QN
Vk[20]=Xk[20]*Sk[20]*Xk[20]
for i in reversed(range(nSteps)):
    Kk[i]=(B*Sk[i+1]*A)/((B*Sk[i+1]*B)+R)
    Sk[i]= Q+(Kk[i]*R*Kk[i])+((A-B*Kk[i])*Sk[i+1]*(A-B*Kk[i]))

ricati_value_lib=np.full_like(Sk, S)
gain_value_lib=np.full_like(Kk, K)
    
    
```

```python
# plot the values
plt.plot(t_span,Sk,label="riccati value")
plt.plot(t_span,Kk,'--',label="LQR gain value")

plt.plot(t_span,ricati_value_lib,'--',label="riccati value using dare")
plt.plot(t_span,gain_value_lib,'--',label="riccati value using dare")

plt.title("comparing gain K and riccati value using manual QP and dare func")
plt.savefig("Comparing  dare func and manual Qp algorithm.png")
plt.legend()
plt.show()
```
